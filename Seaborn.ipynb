```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.relplot(x = "total_bill", y = "tip", data = df)
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df, hue = "smoker")
```

```python
sns.scatterplot(x = "total_bill", y = "tip", data = df, style = "smoker", hue = "time")
```

```python
time = np.arange(500)
value = np.random.randn(500).cumsum()
df = pd.DataFrame({"time": time, "value": value})
df
```

```python
sns.relplot(x = "time", y = "value", data = df, kind = "line", sort = True)
```

```python
tips = (sns.load_dataset("tips"))
tips
```

```python
sns.relplot(x = "total_bill", y = "tip", hue = "smoker", data = tips, kind = "scatter", col = "day")
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "strip")
```

```python
 sns.catplot(x = "day", y = "total_bill", data = tips, kind = "swarm")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="box", palette="Set1")
```

```python
sns.catplot(x="day", y="total_bill", data=tips, kind="violin", palette="Set1")
```

```python
 sns.catplot(x = "smoker", y = "total_bill", data = tips, kind = "swarm")
```

```python
sns.jointplot(x = "total_bill", y = "tip", kind = "reg", data = tips)
```

```python
sns.jointplot(x  = tips.total_bill, y = tips.tip, kind = "kde")
```

```python
sns.pairplot(tips, hue = "smoker", palette= None)
```

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, palette = "rainbow")
```
